```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

def f(x,y,alpha,p):
  return (x-1)**2 + (y-2)**2 + alpha*(abs(x)**p + abs(y)**p)

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X, Y = np.meshgrid(x,y)
alpha = 6
p = 2
Z = f(X,Y,alpha,p)

plt.figure(figsize=(6,6))
plt.contourf(X,Y,Z,20,cmap='coolwarm')
```
